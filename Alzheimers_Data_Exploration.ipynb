```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv('https://data.cdc.gov/resource/hfr9-rurv.csv')
```

```python
df.head()
```

```python
conn = sqlite3.connect('healthcare.db')
```

```python
df.to_sql('alzheimers_data', conn, if_exists='replace', index=False)
```

```python
query1 = 'SELECT * FROM alzheimers_data WHERE locationdesc="New York"'
```

```python
df_query1 = pd.read_sql_query(query1, conn)
df_query1.head()
```

```python
query2 = 'SELECT COUNT(*) FROM alzheimers_data WHERE high_confidence_limit > 40'
```

```python
count_result = pd.read_sql_query(query2, conn)
print(count_result)
```

```python
query3 = 'SELECT low_confidence_limit, AVG(locationid) as av_id FROM alzheimers_data GROUP BY low_confidence_limit'
```

```python
df_query3 = pd.read_sql_query(query3, conn)
df_query3.head()
```

```python
query4 = 'SELECT * FROM alzheimers_data ORDER BY locationid DESC LIMIT 10'
```

```python
df_query4 = pd.read_sql_query(query4, conn)
df_query4
```
